```python
all_dict = {}

for s in df['Tags'].values:
    if isinstance(s, str):
        lst = s.split('|')
        for l in lst:
            if l in all_dict.keys():
                all_dict[l] += 1
            else:
                all_dict[l] = 1

sorted_dict = dict(sorted(all_dict.items(), key=lambda item: -item[1]))

print(list(sorted_dict.keys())[:10]) # top 10
```

```python
tags = list(sorted_dict.keys())[:10]
```
