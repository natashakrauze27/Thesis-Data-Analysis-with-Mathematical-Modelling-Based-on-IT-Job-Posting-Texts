## Построение матрицы релевантности вакансия - фраза из "очищенного" списка требований

Очищенный список - список, из которого удалены фразы с релевантностью меньше 0.1

```python
f = open('req_picked_clean.txt', 'r')
req_picked = f.readlines()
f.close()
data_req= list(data_req)
req_picked = list(map(lambda x: re.sub(pat,' ',x.replace('\n',' ')).lower(), req_picked))
```

```python
def make_substrings(tokens, k=4):
    tokens = tokens.split()
    for i in range(max(len(tokens) - k + 1, 1)):
        yield ' '.join(tokens[i:i + k])
        
def get_relevance_matrix(texts, strings):

    matrix = np.empty((0, len(strings)), float)

    for text in texts:
        ast = base.AST.get_ast(list(make_substrings(text)))
        row = np.array([ast.score(s) for s in strings])
        matrix = np.append(matrix, [row], axis=0)

    return matrix
```

```python
relevance_matrix = get_relevance_matrix(data_req, req_picked)

print("Relevance mairix:")
```

```python
np.savetxt('relevance_matrix_req.txt', relevance_matrix)
```

```python
mask = relevance_matrix>=0.3
mask
```

```python
idx_x_requirements = pd.DataFrame(columns=['ad_id','requirements_refined','req_idxs'])
```

```python
for i in range(len(mask)):
    idx_x_requirements = idx_x_requirements.append({'ad_id': idxs[i],'requirements_refined':np.array(req_picked)[mask[i]],'req_idxs':np.where(relevance_matrix[i]>=0.3)}, ignore_index=True)
    print(np.array(req_picked)[mask[i]])
```
